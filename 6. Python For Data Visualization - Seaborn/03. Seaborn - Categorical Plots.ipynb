## Categorical Plots

```python
import seaborn as sns
import numpy as np
```

```python
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

### Bar Plot

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

### Count Plots

```python
sns.countplot(x='sex',data=tips)
```

### Box Plot (Box-and-Whisker Plot)

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker',palette='coolwarm')
```

### Violin Plot

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker',palette='coolwarm',split=True)
```

### Strip Plot

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',palette='rocket',split=True)
```

### Swarm Plot

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
## Combining swarm plot and violin plots
sns.violinplot(x='day',y='total_bill',data=tips,palette='coolwarm')
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

### Factor Plot

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar',palette='coolwarm')
```
