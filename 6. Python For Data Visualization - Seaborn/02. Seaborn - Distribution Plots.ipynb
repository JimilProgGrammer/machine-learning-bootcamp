# Distribution Plots

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
sns.set()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

## Distplots

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
# plots a histogram and a KDE - kernel density estimation - over the distribution
# to avoid plotting the KDE:
#     sns.distplot(df['column'],kde=False)
```

## Joint Plots

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
# Regression line on scatter plot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

## Pair Plot

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex')
```

```python
sns.pairplot(tips,hue='sex',palette='rocket')
```

## Rug Plot

```python
sns.rugplot(tips['total_bill'])
```

## KDE Plot

```python
sns.kdeplot(tips['total_bill'])
```
