## K-Means Clustering: Unsupervised Learning

```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style("whitegrid")
```

### Generating artificial data with SciKit-Learn

```python
from sklearn.datasets import make_blobs
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

### Exploratory Data Analysis

```python
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='plasma')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
fig, (ax1,ax2) = plt.subplots(1, 2, sharey=True, figsize=(10,6))

ax1.set_title("K Means")
ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap='plasma')

ax2.set_title("Original")
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='plasma')

centers = np.array(kmeans.cluster_centers_)
ax1.scatter(centers[:,0], centers[:,1], marker="x", s=200, color='mediumseagreen')
```
