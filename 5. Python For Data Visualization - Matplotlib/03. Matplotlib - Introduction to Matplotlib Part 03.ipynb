## Introduction to Matplotlib, Part 03

```python
# Imports
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

### Plot Appearance

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='cornflowerblue',linewidth=2,alpha=1,linestyle='-',marker='o',markersize='10'
        ,markerfacecolor='magenta',markeredgewidth=2,markeredgecolor='crimson')
```

### Control over Axis Appearance

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='cornflowerblue',linewidth=2,alpha=1,linestyle='--')

ax.set_xlim([0,2])
ax.set_ylim([0,5])
```
