# Introduction to Matplotlib, Part 02

```python
# Imports Section
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Setup Section
x = np.linspace(0,5,11)
y = x ** 2
```

## Subplots using Object-Oriented Method

```python
# axes will be an array
# subplots() calls add_axes() internally for the number of rows and cols that you have mentioned
fig,axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y,'r--')
axes[0].set_title("X vs Y")

axes[1].plot(y,x,'b-*')
axes[1].set_title("Y vs X")

# To avoid overlapping
plt.tight_layout()
```

## Fig Size and DPI

```python
# figsize argument takes a tuple in the form (width,height)
fig = plt.figure(figsize=(4,2), dpi=100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(3,3),dpi=100)

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

## Save figure

```python
fig.savefig("Sample1.jpg",dpi=200)
```

## Legend

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label="X vs X^2")
ax.plot(x,x**3,label="X vs X^3")

ax.legend(loc=0)
```
