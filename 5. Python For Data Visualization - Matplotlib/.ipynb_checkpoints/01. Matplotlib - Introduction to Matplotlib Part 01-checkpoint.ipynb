# Introduction to Matplotlib, Part 1

```python
# Imports section
import matplotlib.pyplot as plt
import numpy as np
```

```python
# This line allows you to view the generated plots inside the notebook
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

## Functional Method to Plot graphs

```python
# Functional Method
plt.plot(x,y,'blueviolet')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title("X Vs Y")
# plt.show()
```

## Multi-plots on same canvas

##### plt.subplot(rows,cols,cell)
##### rows - number of rows on the canvas
##### cols - number of cols on the canvas
##### cell - the cell you are currently plotting or editing

```python
# 1 row, 2 columns, 1st cell
plt.subplot(1,2,1)
plt.plot(x,y,'r')

# 1 row, 2 columns, 2nd cell
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

## Object-Oriented Method
#### Using the OO method, you use matplotlib to obtain figure objects and then call methods off of it

```python
# The figure object is kind of like a blank canvas
fig = plt.figure()

# Adding axes to the canvas
# add_axes([left,bottom,width,heigh]) - percentage (between 0 to 1) of the canvas that is to be covered
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)

# Set Labels and title
axes.set_xlabel('X Label')
axes.set_ylabel("Y Label")
axes.set_title("X VS Y")
```

## Multi-plots using Axes

```python
fig = plt.figure()

# first axis
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])

# the second axis starts 20% from left, 50% from bottom, covers 40% width and 30% height
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
```

```python
fig = plt.figure()

# first axis
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])

# the second axis starts 20% from left, 50% from bottom, covers 40% width and 30% height
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

# Plot on the first axis
axes1.plot(x,y,'magenta')
# Plot on the second axis
axes2.plot(y,x,'cornflowerblue')

# Set xlabel, ylabel, title for axes1
axes1.set_xlabel("A1 X Label")
axes1.set_ylabel("A1 Y Label")
axes1.set_title("A1 X vs. Y")

# Set xlabel, ylabel, title for axes2
axes2.set_xlabel("A2 Y Label")
axes2.set_ylabel("A2 X Label")
axes2.set_title("A2 Y vs. X")
```
