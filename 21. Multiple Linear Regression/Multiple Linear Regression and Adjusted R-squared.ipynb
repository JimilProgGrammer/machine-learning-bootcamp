## Multiple Linear Regression

### Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
sns.set()
```

### Load the data

```python
data = pd.read_csv('sat_random.csv')
data.drop('Unnamed: 0', axis=1, inplace=True)
data.head()
```

### Creating the regression

```python
y = data['HS GPA']
x1 = data[['SAT Score', 'Random']]
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
```

```python
results.summary()
```

```python
x
```

### Making Predictions

```python
new_data = pd.DataFrame({'const':1,'SAT':[1700,1670],'Random':[1,3]})
new_data = new_data[['const','SAT','Random']]
new_data
```

```python
predictions = results.predict(new_data)
predictions
```
