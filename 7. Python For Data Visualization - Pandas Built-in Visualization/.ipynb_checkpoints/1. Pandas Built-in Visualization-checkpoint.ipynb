## Visualization with Pandas

```python
# Imports section
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
# Setting default styles
sns.set_style('whitegrid')
sns.set_palette('rocket')
```

```python
# Prepare data
df1 = pd.read_csv('df1', index_col=0)
df1['Year'] = df1.index
df1.head()
```

```python
df2 = pd.read_csv('df2')
df2.head()
```

### Histograms

```python
df1['A'].hist(bins=30)
```

```python
df1['A'].plot(kind='hist',bins=30)
```

```python
df1['A'].plot.hist(bins=30)
```

### Area Plots

```python
df2.plot.area(alpha=0.4)
```

### Bar Plots

```python
df2.plot.bar(stacked=True)
```

### Line Plots

```python
df1.plot.line(x='Year', y='B',figsize=(12,3),lw=1)
```

### Scatter Plots

```python
df1.plot.scatter(x="A",y="B",c="C",cmap='rocket')
```

```python
# To show scatter plot by size
df1.plot.scatter(x="A", y="B", s=df1['C']*100)
```

### Box Plots

```python
df2.plot.box()
```

```python
df = pd.DataFrame(np.random.randn(1000,2), columns=['a','b'])
df.head()
```

### Hexbin Plots

```python
df.plot.hexbin(x='a',y='b',gridsize=25,figsize=(10,8),cmap='rocket')
```

### KDE Plots

```python
df2['b'].plot.kde() # or plot.density()
```

```python
df2.plot.kde(cmap='coolwarm')
```
